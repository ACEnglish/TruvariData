```python
import joblib
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = joblib.load("../stats/inter_merge_size_summary.jl")
```

```python
data.head()
```

```python
data = data[data["szbin"] != "[0,50)"]
```

```python
plt.figure(dpi=180)
p = sb.catplot(kind='bar', 
               data=data, 
               col='merge',
               row="ref", 
               x="szbin", 
               y='count', 
               hue='svtype',
               hue_order=["DEL", "INS"],
               col_order=['exact', 'truvari', 'jasmine', 'naive', 'survivor'])

for axes in p.axes.flat:
    _ = axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
    axes.grid(b=True,which='major',axis='y', linewidth=.8)
    axes.set_axisbelow(True)
```
